Yapmak İstediğim; İki Excel dosyasını karşılaştırılarak **Ana_Tablo**daki verileri güncellemek
**Güncel_Kurum_Sayıları** tablosunda olup, **Ana_Tablo**da yoksa olmayan kurumun Ana_tabloya **eklenmesi** ve **Durumu** başlığına **Yeni Açılış (23.03.2024)** olarak not eklenmesi,
**Güncel_Kurum_Sayıları** tablosunda **olmayıp**, **Ana_tablo**da var ise **Ana_tablodan bunların silinmesi** ve mesaj olarak 'silinen kurumun ilçesi ve okul adının (Örnek;‘Sarıca İlçesi-Sarıca İlkokulu Kapatılmıştır’) şeklinde **bildirilmesi**

```python
import pandas as pd
df_ana = pd.read_excel("Ana_Tablo.ods")
df_ana.head()
```

```python
df_guncel = pd.read_excel("Güncel_Kurum_Sayıları.ods")
df_guncel.head()
```

```python
df_ana["Kurum_Kod_Ad"] = df_ana["KURUM_KODU"].astype(str) + df_ana["KURUM_ADI"]
df_ana = df_ana.set_index("Kurum_Kod_Ad")
df_ana.head()
```

```python
df_guncel["Kurum_Kod_Ad"] = df_guncel["KURUM_KODU"].astype(str) + df_guncel["KURUM_ADI"]
df_guncel = df_guncel.set_index("Kurum_Kod_Ad")
df_guncel.head()
```

```python
df_ana_indeks = df_ana.index
df_ana_indeks
```

```python
df_guncel_indeks = df_guncel.index
df_guncel_indeks
```

```python
ekle = []

for indeks in df_guncel_indeks:
    if indeks not in df_ana_indeks:
        ekle.append(indeks)

ekle
```

```python
sil = []

for indeks in df_ana_indeks:
    if indeks not in df_guncel_indeks:
        sil.append(indeks)

sil
```
